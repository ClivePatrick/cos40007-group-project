```python
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('clean_data.csv')
df.head()
```

```python
avg_latency = np.mean(df[['svr1', 'svr2', 'svr3', 'svr4']], axis=1)
```

```python
# Create a Scatter plot of the latitude and longitude points, coloured by the average latency
im = plt.scatter(x=df['latitude'],
                 y=df['longitude'],
                 s=0.25,
                 vmin=np.min(avg_latency),
                 vmax=np.max(avg_latency),
                 c=avg_latency,
                 alpha=0.5,
                 cmap=cm.plasma)
plt.colorbar(im)
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.axis([-37.800, -37.700, 144.750, 144.850])
plt.show()
```

```python
fig, ax = plt.subplots(1, 3, figsize=(9, 3), sharex=True, sharey=True)

features = [
    ('Average Latency (sec)', avg_latency),
    ('Upload (mbits/sec)', df['upload_bitrate_mbits/sec']),
    ('Download (mbits/sec)', df['download_bitrate_rx_mbits/sec'])
]

for index, feature in enumerate(features):
    title, column = feature
    im = ax[index].scatter(x=df['latitude'],
                           y=df['longitude'],
                           s=0.25,
                           vmin=np.min(column),
                           vmax=np.max(column),
                           c=column,
                           alpha=0.5,
                           cmap=cm.plasma)
    ax[index].set_title(title)
    ax[index].set_xlim(-37.800, -37.700)
    ax[index].set_ylim(144.750, 144.850)

plt.show()
```
